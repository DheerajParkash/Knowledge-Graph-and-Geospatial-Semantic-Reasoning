```python
import pandas as pd

# Load the dataset (replace 'your_dataset.csv' with the actual file path)
dataset_path = 'whc-sites-2024.xlsx'
data = pd.read_excel(dataset_path)

# Select the required columns and filter rows where states_name_en is "France"
columns_to_keep = [
    "unique_number", "iso_code", "name_en", "longitude", "latitude", "states_name_en", 
    "region_en"
]
filtered_data = data[data["states_name_en"] == "France"][columns_to_keep]

# Add a new column "coordinates" in Point(longitude, latitude) format
filtered_data["coordinates"] = filtered_data.apply(lambda row: f"Point({row['longitude']} {row['latitude']})", axis=1)

# Save the filtered data to a CSV file
output_path = 'unhsite.csv'
filtered_data.to_csv(output_path, index=False)

print(f"Filtered data saved to {output_path}")
```
